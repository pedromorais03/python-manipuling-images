```python
import cv2
from google.colab.patches import cv2_imshow

imagem = cv2.imread('neto.jpeg')
for y in range(0, imagem.shape[0]):
  for x in range(0, imagem.shape[1]):
    (azul,verde,vermelho) = imagem[y,x]
    imagem[y,x] = (0,0,2+2*vermelho*azul%256)
cv2_imshow(imagem)
```
