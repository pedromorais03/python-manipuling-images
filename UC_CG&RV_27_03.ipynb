```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('img01.jpeg')
cv2_imshow(image)
```

```python
x = 25
y = 72
(azul, verde, vermelho) = image[x, y]
print(f'Vermelho: {vermelho} \nVerde: {verde} \nAzul: {azul}')
```

```python
altura = image.shape[0]
largura = image.shape[1]
profundidade = image.shape[2]

print(f'Altura: {altura} \nLargura: {largura} \nProfundidade: {profundidade}')

# A profundidade é 3, pq cada cor está em uma camada
```

```python
# Separando os valores de cada cor com o split
(mAzul, mVerde, mVermelho) = cv2.split(image)
print(f'Matriz Azul: {mAzul[25][72]} \nMatriz Verde: {mVerde[25][72]} \nMatriz Vermelho: {mVermelho[25][72]}')
```

```python
# Essas matrizes não possuem profundidade, são matrizes bidimensionais
altura = mVermelho.shape[0]
largura = mVermelho.shape[1]

print(f'Altura: {altura} \nLargura: {largura}')
```

```python
# Se eu mostrar a imagem com uma das 3 matrizes, o programa irá colocar o mesmo valor de Azul para o Verde e para o Vermelho
cv2_imshow(mAzul)
```

```python
import numpy as np

# Criando uma matriz de 0's, do tamanho da Altura x Largura da imagem.
zeros = np.zeros(image.shape[0:2], dtype="uint8")

# Exibindo a imagem em uma cor apenas
cv2_imshow(cv2.merge([zeros, zeros, mVermelho]))
```

```python
cv2_imshow(cv2.merge([zeros, mVerde, zeros]))
```

```python
cv2_imshow(cv2.merge([mAzul, zeros, zeros]))
```

```python
cv2_imshow(cv2.merge([mAzul, zeros, mVermelho]))
cv2_imshow(cv2.merge([mAzul, mVerde, zeros]))
cv2_imshow(cv2.merge([zeros, mVerde, mVermelho]))
```
