```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
imagem = cv2.imread("22_alterado.png")
cv2_imshow(imagem)
```

```python
## aq os valores nao sao (x,y), sao (y,x)
altura = imagem.shape[0]
largura = imagem.shape[1]
```

```python
print(f'A altura √© {altura} e a largura √© {largura}')
```

```python
## aq os valores nao sao (x,y), sao (y,x) e os valores da cor sao azul, verde e vermelho
(azul, verde, vermelho) = imagem[50, 63]
print(f'Vermelho: {vermelho} - Verde: {verde} - Azul: {azul}')
```

```python
imagem[65, 17] = (255,0,0)
imagem[24, 69] = (255,0,0)
imagem[24, 7] = (255,0,0)
imagem[140, 25] = (255,0,0)
cv2_imshow(imagem)
```
